```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

C = [2*np.sin(3*np.pi*t) for t in np.linspace(0,1, 1000)]

H = [4*np.sin(8*np.pi*t) for t in np.linspace(0,3, 1000)]

I = [6*np.sin(9*np.pi*t) for t in np.linspace(0,4, 1000)]

output=[i+J+K for i,J,K in zip(C,H,I)]

plt.plot(output)

Chi=np.fft.fft(output)
plt.plot(output)
print(output)
plt.plot(Chi)
plt.plot(Chi)
plt.show()
```
